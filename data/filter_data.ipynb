```python
import pandas as pd
```

```python
data = pd.read_csv('results.csv', error_bad_lines=False)
```

```python
data
```

```python
# Merge gold, bronze, and silver into one row.
data = pd.merge(
    pd.merge(data[data['Medal'] == 'G'],
             data[data['Medal'] == 'S'], 
             how='inner',
             on=['Gender','Event','Location','Year'],
             suffixes=('', '_S')),
    data[data['Medal'] == 'B'],
    how='inner',
    on=['Gender','Event','Location','Year'],
    suffixes=('', '_B'))
data
```

```python
# List of ('Event' filter for dataframe, outfile path, race distance) pairs.
events_to_filter = [
    ('100M Men', '100m_men.csv', 100),
    ('100M Women', '100m_women.csv', 100),
    ('200M Men', '200m_men.csv', 200),
    ('200M Women', '200m_women.csv', 200),
    ('400M Men', '400m_men.csv', 400),
    ('400M Women', '400m_women.csv', 400),
]
```

```python
def augment_df_rows(df):
    """Copies df and adds (0, 0) time and distances rows."""
    df_zero = df.copy()
    df_zero['Time'] = 0
    df_zero['Distance'] = 0
    return df_zero.append(df, ignore_index=True)
```

```python
for event, outfile_path, race_distance in events_to_filter:
    filtered_data = data[(data['Event'] == event) & (data['Medal'] == 'G')]
    filtered_data['Distance'] = race_distance
    filtered_data.rename(index=str, columns={"Result": "Time"}, inplace=True)
    print('Event %s contains %d rows (before augmenting).' % (event, len(filtered_data)))
    filtered_data = augment_df_rows(filtered_data)
    filtered_data.to_csv(outfile_path, index=False)
```
